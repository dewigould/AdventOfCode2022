## Part 1

```python
import numpy as np
```

```python
with open('inputday5.txt') as f:
    lines = [line for line in f]   
```

```python
lines
```

```python
class Stack():
    """A class for a stack
    """
    
    def __init__(self,crates):
        self.crates = Stack.remove_blanks(crates)
        self.top_crate = crates[-1]
        
    def remove_blanks(input_list):
        return [i for i in input_list if i != " "]
        
    def get_top_N_crates(self,N):
        return self.crates[len(self.crates)-N:]
    
    def remove_N_crates(self,N):
        self.crates = self.crates[:-N]
        
    def add_crates(self,crates_to_add,reversal):
        if reversal:
            crates_to_add.reverse()
        self.crates = self.crates + crates_to_add
        

         
            
class CollectionOfStacks():
    """A class for the full collection of stacks
    """
    
    def __init__(self,list_of_stacks):
        self.list_of_stacks = list_of_stacks
        
    def update_with_instruction(self,instruction,with_reversal):
        crates_to_move = self.list_of_stacks[instruction.start_col].get_top_N_crates(instruction.num)
        #remove these from col 1
        self.list_of_stacks[instruction.start_col].remove_N_crates(instruction.num)
        #add these to col 2
        self.list_of_stacks[instruction.end_col].add_crates(crates_to_move,with_reversal)
        
        
    def show_stack_status(self):
        return([i.crates for i in self.list_of_stacks])
    
    def return_top_crates(self):
        return[i.get_top_N_crates(1) for i in self.list_of_stacks]
        

class Instruction():
    """A class for an instruction
    """
    
    def __init__(self,how_many,start_col,end_col):
        self.num = how_many
        self.start_col = start_col-1 #Python indices
        self.end_col = end_col-1 #Python indices
        


def get_instruction_data(instruction_string):
    res = [int(i) for i in instruction_string.split() if i.isdigit()]
    return(Instruction(res[0],res[1],res[2]))
```

```python
num_stacks = 9
crate_input=lines[:num_stacks-1]
crate_input.reverse()
```

```python
# Horrible hard coding to add get crate orderings into Stacks class
formatted_crate_list=np.array([[crate_list[j] for j in (1,5,9,13,17,21,25,29,33)] for crate_list in crate_input])
transposed_crate_list=np.transpose(formatted_crate_list)
stacks = [Stack(list(crate_list)) for crate_list in transposed_crate_list ]
SetofStacks = CollectionOfStacks(stacks)
```

```python
SetofStacks.show_stack_status()
```

```python
instructions_list=lines[num_stacks+1:]
```

```python
for line in instructions_list:
    instruction = get_instruction_data(line)
    SetofStacks.update_with_instruction(instruction,with_reversal=True)
```

```python
SetofStacks.show_stack_status()
```

```python
SetofStacks.return_top_crates()
```
